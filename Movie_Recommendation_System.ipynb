```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies = pd.read_csv("/content/movies.csv")
print(movies)
```

```python
print(movies.head())
```

```python
print("Available Genres:")
genres = set()

for genre_list in movies["genres"]:
    for genre in genre_list.split("|"):
        genres.add(genre)

print(sorted(genres))
```

```python
user_genre = input("\nEnter your favorite genre: ")
```

```python
recommended_movies = movies[
    movies["genres"].str.lower().str.contains(user_genre.lower(), na=False)
]
```

```python
recommended_movies = recommended_movies[["movieId", "title", "genres"]].head(10)

print("\nRecommended Movies:\n")

if recommended_movies.empty:
    print("Sorry, no movies found for this genre.")
else:
    print(recommended_movies.to_string(index=False))
```
